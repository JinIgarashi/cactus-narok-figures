## Figure 10 Total Annualised Cost per Household (TACH) (Int$ (2020)) for whole sanitation systems for Narok town (box plots) compared to the CACTUS database (bars) in September 2022  


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.text import Text
import textwrap
import matplotlib.patches as mpatches
from matplotlib.path import Path
import numpy as np
```

```python
# variables
path_excel = r'data.xlsx'
sheet_name = 'full_lifecycle_cost_compare'
file_path = r'images/fig10_fulllifecycle_cost_comparison.pdf'
dpi = 700

plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.size'] = 14

fsm1_tachs = [345,438]
fsm1_pos = 2
fsm2_tachs = [380,473]
fsm2_pos = 5
sewer_tachs = [623, 6022]
sewer_pos = 12
```

```python
df = pd.read_excel(
    path_excel, 
    sheet_name=sheet_name,
)
```

```python
fig, axs = plt.subplots(nrows=2, figsize=(15, 8), sharex='col',
                       gridspec_kw={'height_ratios': (1,2)} )

fig.patch.set_facecolor('white') # 

hatches = ['///', '...', 'xxx'] 
labels = []
# labels.append("Full lifecycle cost")

for index, ax in enumerate(axs):
    bottom = None
    total = np.zeros(len(df))
    for i, column in enumerate(df.columns[1:]):
        bar = ax.bar(
            df.iloc[:, 0], 
            df[column], 
            width=0.3,
            color='w', 
            edgecolor='k', 
            bottom=bottom, 
            label=column, 
            hatch=hatches[i % len(hatches)]
        )
        bottom = df[column] if bottom is None else bottom + df[column]
        total += df[column]
        if index == 1:
            labels.append(column)

        # 
        for j, value in enumerate(df[column]): 
            if np.isnan(value):
                continue
            if index == 0:
               if bottom[j] < 5000:
                continue
            else:
                if bottom[j] > 5000:
                    continue
            y = bottom[j] - 20
            if j == 5 and i == 1:
                y = bottom[j] + 5
            ax.text(
                j + 0.2, 
                y, 
                int(round(value)), 
                ha='left', 
                va='center', 
                fontsize=9,
                bbox=dict(boxstyle='round, pad=0.1', facecolor='w', edgecolor='none', alpha=0.7))
    
    # add vertial line to separate FSM and Sewage
    for i in range(len(df)-1):
        if i == 5:
            ax.axvline(i + 0.6, color='grey', linestyle='-', alpha=1)
    
    # 
    for j, value in enumerate(total):
        if np.isnan(value):
                continue
        if index == 0:
            if value < 5000:
                continue
        else:
            if value > 5000:
                continue

        ax.plot(
            j, 
            value, 
            color='k', 
            linestyle='', 
            marker='o', 
            markersize=10, 
            label='Full lifecycle cost')
        
        ax.text(
            j,
            value + 40,
            int(round(value)),
            ha='center',
            va='center',
            fontsize=12,
            bbox=dict(boxstyle='round, pad=0.1', facecolor='w', edgecolor='none', alpha=0.7)
        )

# make boxplots for the range of Narok town's cost
#FSM - pitlatrine
bp1=  axs[1].boxplot(
    fsm1_tachs,
    positions= [fsm1_pos]
)
# make median line black
for median in bp1['medians']:
    median.set_color('black')
    
q1, q3 = min(fsm1_tachs), max(fsm1_tachs)
axs[1].text(fsm1_pos, q1 - 50, f"{q1:.0f}", ha='center', va='center', fontsize=12)
axs[1].text(fsm1_pos, q3 + 50, f"{q3:.0f}", ha='center', va='center', fontsize=12)

#FSM - septic tank
bp2 = axs[1].boxplot(
    fsm2_tachs,
    positions= [fsm2_pos]
)
# make median line black
for median in bp2['medians']:
    median.set_color('black')
    
q1, q3 = min(fsm2_tachs), max(fsm2_tachs)
axs[1].text(fsm2_pos, q1 - 50, f"{q1:.0f}", ha='center', va='center', fontsize=12)
axs[1].text(fsm2_pos, q3 + 50, f"{q3:.0f}", ha='center', va='center', fontsize=12)

#Sewer
for i in range(2):
    axs[i].boxplot(
        sewer_tachs,
        positions= [sewer_pos]
    )
    q1, q3 = min(sewer_tachs), max(sewer_tachs)
    if i > 0:
        axs[i].text(sewer_pos, q1 - 50, f"{q1:.0f}", ha='center', va='center', fontsize=12)
    if i == 0:
        axs[i].text(sewer_pos, q3 + 20, f"{q3:.0f}", ha='center', va='center', fontsize=12)
        
axs[0].axhline(y=6022,xmin=0.5, xmax=0.95, color='grey', linestyle='-.', alpha=1)
axs[0].axhline(y=623,xmin=0.5, xmax=0.95, color='grey', linestyle='-.', alpha=1)
axs[1].axhline(y=623,xmin=0.5, xmax=0.95, color='grey', linestyle='-.', alpha=1)
axs[1].axhline(y=438,xmin=0, xmax=0.2, color='grey', linestyle='-.', alpha=1)
axs[1].axhline(y=345,xmin=0, xmax=0.2, color='grey', linestyle='-.', alpha=1)
axs[1].axhline(y=473,xmin=0.25, xmax=0.43, color='grey', linestyle='-.', alpha=1)
axs[1].axhline(y=380,xmin=0.25, xmax=0.43, color='grey', linestyle='-.', alpha=1)

# create legend with reversed order
axs[0].legend(loc='upper left')
handles, labels = axs[0].get_legend_handles_labels()
axs[0].legend(reversed(handles), reversed(labels), loc='upper left')

xtick_labels = [ '\n'.join(textwrap.wrap(label, 20)) for label in df.iloc[:, 0] ]
axs[1].set_xticks(range(len(xtick_labels)))
axs[1].set_xticklabels(xtick_labels, rotation=90, ha='center', fontsize=12)

fig.subplots_adjust(hspace=0.0)  

# 
axs[1].set_ylim(0,1250)  #  400
axs[1].set_yticks(np.arange(0,1250,250))

# 
axs[0].set_ylim(5800,6100)  #  200
axs[0].set_yticks(np.arange(5900,6100,100))

# 
axs[1].spines['top'].set_visible(False)

# X
axs[0].spines['bottom'].set_visible(False)
axs[0].tick_params(axis='x', which='both', bottom=False, labelbottom=False) 

## 
d1 = 0.02 # X
d2 = 0.03 # 
wn = 21   # 

pp = (0,d2,0,-d2)
px = np.linspace(-d1,1+d1,wn)
py = np.array([1+pp[i%4] for i in range(0,wn)])
p = Path(list(zip(px,py)), [Path.MOVETO]+[Path.CURVE3]*(wn-1))

line1 = mpatches.PathPatch(p, lw=4, edgecolor='black',
                          facecolor='None', clip_on=False,
                          transform=axs[1].transAxes, zorder=10)

line2 = mpatches.PathPatch(p,lw=3, edgecolor='white',
                           facecolor='None', clip_on=False,
                           transform=axs[1].transAxes, zorder=10,
                           capstyle='round')

a = axs[1].add_patch(line1)
a = axs[1].add_patch(line2)

# axs[0].set_title('The comparison of full lifecycle cost between Narok town and the CACTUS database', loc='center', x=0.5, y=1.03, fontsize=14)
axs[1].set_ylabel('Total Annualised Cost per Household (TACH) Int$ (2020)', fontsize=14)

axs[0].text(4, 6070, 'Fecal Sludge Management\n(FSM) System', fontsize=14, ha='center', va='top', bbox=dict(facecolor='lightgray', edgecolor='none', alpha=0.8, boxstyle='round'))
axs[0].text(6.5, 6070, 'Sewer System', fontsize=14, ha='center', va='top', bbox=dict(facecolor='lightgray', edgecolor='none', alpha=0.6, boxstyle='round'))

plt.savefig(file_path, format="pdf", dpi=dpi, transparent=True, bbox_inches='tight')
plt.show()
```
